### Gráficos e Insights

Gráfico 1 - Relação entre modelo de trabalho e horas trabalhadas

Este gráfico mostra a média de horas semanais trabalhadas para cada categoria de modelo de trabalho (WORKCLASS)

px.bar(
    data_graph_workclass,
    x='classe_trabalhadora_traduzida',
    y='horas_por_semana',
    title='Relacão entre Modelo de Trabalho e horas de Trabalho',
    color='classe_trabalhadora_traduzida',
    labels={
        'classe_trabalhadora_traduzida': 'Modelo de Trabalho',
        'horas_por_semana': 'Horas Trabalhadas'
    }
)

 - Insight: Há diferenças significativas no tempo médio trabalhado com base no modelo de trabalho. Por exemplo, pessoas em trabalho autônomo podem ter horários mais flexíveis, enquanto empregados privados tendem a seguir jornadas padronizadas.

Gráfico 2 - Boxplot das Horas Trabalhadas por semana 

O boxplot revela a distribuição das horas trabalhadas, incluindi possíveis outliers.

px.box(data['hours-per-week'])

 -  Insight: A maioria dos indivíduos trabalha cerca de 40 horas por semana, mas há outliers que trabalham bem mais ou menos do que isso. Esses valores extremos foram tratados com um filtro baseado no desvio padrão.

Gráfico 3 - Disperção entre Idade, Horas, Trabalhadas e Renda

A relação entre idade, horas trabalhadas por semana e faixa de renda(INCOME).
